```python
import pandas as pd

### CRISPR knockout screen across 1000+ cell lines

df_crispr = pd.read_csv('CRISPRGeneEffect.csv',index_col=0)
df_exp = pd.read_csv('OmicsExpressionProteinCodingGenesTPMLogp1.csv',index_col=0)
```

```python
df_crispr
```

```python
df_crispr.columns = [s[0:s.index('(')-1] for s in df_crispr.columns]
df_exp.columns = [s[0:s.index('(')-1] for s in df_exp.columns]
```

```python
## Get cell lines that are common in CRISPR and RNAseq data

ind_ls = list(set(df_crispr.index)&set(df_exp.index))
```

```python
import matplotlib.pyplot as plt
import numpy as np

gene = 'SOX10'

plt.figure(figsize=(3,2),dpi=300)

plt.rcParams["font.size"] = 8
plt.rcParams["font.family"] = "Helvetica"

plt.scatter(df_exp.loc[ind_ls,gene],df_crispr.loc[ind_ls,gene],s=10, alpha=0.5, color='blue')

plt.xlabel('Expression of %s'%gene)
plt.ylabel('CRISPR score of %s'%gene)

plt.show()
```

```python
df_cell = pd.read_csv('Model.csv')
```

```python
df_cell
```

```python
set(df_cell['OncotreeLineage'])
```

```python
cancer_ls = ['Bone','Bowel','Breast','CNS/Brain','Esophagus/Stomach', 'Head and Neck','Kidney',
             'Liver','Lung','Lymphoid','Myeloid', 'Pancreas','Prostate', 'Skin','Soft Tissue']

colors  = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'brown', 'gray', 
           'black', 'cyan', 'magenta', 'olive', 'teal', 'navy', 'maroon']
```

```python
# cancertype = 'Lung'
# df_breast = df_cell[df_cell['OncotreeLineage']==cancertype]
# cell_ls = [c for c in list(df_breast['ModelID']) if c in ind_ls]

gene = 'SOX10'

plt.figure(figsize=(4,3),dpi=300)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.rcParams["font.size"] = 8
plt.rcParams["font.family"] = "Helvetica"

#plt.scatter(df_exp.loc[ind_ls,gene],df_crispr.loc[ind_ls,gene],s=10, alpha=0.5, color='blue')

color_ind = 0
for cancertype in cancer_ls:
    df_cancer = df_cell[df_cell['OncotreeLineage']==cancertype]
    cell_ls = [c for c in list(df_cancer['ModelID']) if c in ind_ls]
    plt.scatter(df_exp.loc[cell_ls,gene],df_crispr.loc[cell_ls,gene],s=6, alpha=0.5, color=colors[color_ind],label=cancertype)
    color_ind += 1

plt.xlabel('Expression of %s'%gene)
plt.ylabel('CRISPR score of %s'%gene)
plt.legend(fontsize='x-small',loc='lower left',ncol=2)

plt.show()
```

```python
df_mut = pd.read_csv('OmicsSomaticMutations.csv')
```

```python
df_mut
```

```python
df_mut.index = list(df_mut['ModelID'])
```

```python
df_mut['MutationType'] = df_mut['HugoSymbol']+'_'+df_mut['ProteinChange']
```

```python
## Get cell lines that are common in CRISPR and mutation data

ind_mut = list(set(df_crispr.index)&set(df_mut.index))
```

```python
import matplotlib.pyplot as plt
import numpy as np

gene = 'KRAS'; mut = 'G12D'
muttype = gene+'_p.'+mut
mut_cell = df_mut[df_mut['MutationType']==muttype]['ModelID']
df_crispr['MutLabel'] = [1 if c in mut_cell else 0 for c in df_crispr.index]

plt.figure(figsize=(3,2),dpi=300)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.rcParams["font.size"] = 8
plt.rcParams["font.family"] = "Helvetica"

cs_ls1 = df_crispr[df_crispr['MutLabel']==1]['KRAS']
cs_ls2 = df_crispr[df_crispr['MutLabel']==0]['KRAS']

plt.boxplot([cs_ls1,cs_ls2],labels=['WithMut','WitoutMut'],showfliers=False)
plt.ylabel('CRISPR score for %s'%gene)
plt.savefig('KRAS_Boxplot.png',dpi=300,facecolor=None,bbox_inches='tight')
plt.show()
```

```python
#cancertype = 'Lung'

plt.figure(figsize=(15,8),dpi=300)
plt.rcParams["font.size"] = 8


n = 1
for cancertype in cancer_ls:
    df_cancer = df_cell[df_cell['OncotreeLineage']==cancertype]
    cell_ls = [c for c in list(df_cancer['ModelID']) if c in df_crispr.index]
    df_sel = df_crispr.loc[cell_ls,:]

    cs_ls1 = df_sel[df_sel['MutLabel']==1]['KRAS']
    cs_ls2 = df_sel[df_sel['MutLabel']==0]['KRAS']
    
    plt.subplot(3,5,n)
    
    plt.boxplot([cs_ls1,cs_ls2],labels=['WithMut','WitoutMut'],showfliers=False)
    plt.ylabel('CRISPR score in %s'%cancertype)
    n += 1

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```
