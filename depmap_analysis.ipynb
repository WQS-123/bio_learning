```python
## Gene expression (RNAseq) in different cell lines
### Gene expression TPM values of the protein coding genes uing log2 transformation 
### with a pseudo-count of 1; log2(TPM+1).

# 1.Count the number of reads mapped to each gene (raw counts).
# 2.Normalize the raw counts by the gene length (usually measured in kilobases, or kb) to account for differences in gene lengths. This step produces the Reads Per Kilobase (RPK) values.
# 3.Sum the RPK values for all genes to get the total RPK.
# 4.Divide each gene's RPK value by the total RPK and multiply by 1,000,000 to obtain the TPM values.
import pandas as pd
df_exp = pd.read_csv('OmicsExpressionProteinCodingGenesTPMLogp1.csv',index_col=0)
```
